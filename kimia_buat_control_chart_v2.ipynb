### Nama Parameter

```python
row_num = parameter_row_idx
col_idx = days_col_idx - 1

for i in hema:
    cell = alphabetical_column[col_idx] + str(row_num)
    #cell = 'A' + str(row_num)
    data_ws[cell] = i
    data_ws[cell].border = thin_border

    ###################### blank area to fill
    for j in range(days_col_idx, data_sum_idx):
        cell = alphabetical_column[j] + str(row_num)
        data_ws[cell].border = thin_border

    ###################### data_sum
    data_sum_i = 0
    x_cell = 'AA'
    SD_cell = 'AA'
    for d in data_sum:
        cell = alphabetical_column[data_sum_idx + data_sum_i] + str(row_num)
        #print(cell)
        if data_sum_i == 0: # sum
            formula = '=SUM(' + alphabetical_column[days_col_idx] + str(row_num) + ':' + alphabetical_column[days_col_idx_end] + str(row_num) + ')'
        elif data_sum_i == 1: # n
            formula = '=COUNTA(' + alphabetical_column[days_col_idx] + str(row_num) + ':' + alphabetical_column[days_col_idx_end] + str(row_num) + ')'
        elif data_sum_i == 2: # x
            formula = '=AVERAGE(' + alphabetical_column[days_col_idx] + str(row_num) + ':' + alphabetical_column[days_col_idx_end] + str(row_num) + ')'
            x_cell = cell
        elif data_sum_i == 3: # SD
            formula = '=STDEV(' + alphabetical_column[days_col_idx] + str(row_num) + ':' + alphabetical_column[days_col_idx_end] + str(row_num) + ')'
            SD_cell = cell
        elif data_sum_i == 4: # CV
            formula = '=' + SD_cell + '*100%/' + x_cell
        elif data_sum_i == 5: # -3SD
            formula = '=' + x_cell + '-(3*' + SD_cell + ')'
        elif data_sum_i == 6: # -2SD
            formula = '=' + x_cell + '-(2*' + SD_cell + ')'
        elif data_sum_i == 7: # -1SD
            formula = '=' + x_cell + '-(1*' + SD_cell + ')'
        elif data_sum_i == 8: # 1SD
            formula = '=' + x_cell + '+(1*' + SD_cell + ')'
        elif data_sum_i == 9: # 2SD
            formula = '=' + x_cell + '+(2*' + SD_cell + ')'
        elif data_sum_i == 10: # 3SD
            formula = '=' + x_cell + '+(3*' + SD_cell + ')'

        data_ws[cell] = formula
        data_ws[cell].border = thin_border
        data_sum_i = data_sum_i + 1

    ###################### vendor_sum
    vendor_sum_i = 0
    x_cell = vendor_x_cell
    SD_cell = vendor_SD_cell
    for d in vendor_sum:
        cell = alphabetical_column[vendor_sum_idx + vendor_sum_i] + str(row_num)
        #print(cell)
        if vendor_sum_i == 0: # x
            #formula = '=' + x_cell + str(row_num)
            formula = '=IF(OR(ISBLANK(' + new_vendor_x_cell + str(row_num) + '),' + new_vendor_x_cell + str(row_num) + '=0),(' + range_kiri_cell + str(row_num) + '+' + range_kanan_cell + str(row_num) + ')/2,' + new_vendor_x_cell + str(row_num) + ')'
            #print(formula)
        elif vendor_sum_i == 1: # SD
            #formula = '=' + SD_cell + str(row_num)
            formula = '=IF(OR(ISBLANK(' + new_vendor_SD_cell + str(row_num) + '),' + new_vendor_SD_cell + str(row_num) + '=0),(' + range_kanan_cell + str(row_num) + '-' + vendor_x_cell + str(row_num) + ')/' + str(range_multiplication) + ',' + new_vendor_SD_cell + str(row_num) + ')'
            #print(formula)
        #if vendor_sum_i == 0: # x
            #formula = '=' + x_cell + str(row_num)
        #elif vendor_sum_i == 1: # SD
            #formula = '=' + SD_cell + str(row_num)
        elif vendor_sum_i == 2: # -3SD
            formula = '=' + x_cell + str(row_num) + '-(3*' + SD_cell + str(row_num) + ')'
        elif vendor_sum_i == 3: # -2SD
            formula = '=' + x_cell + str(row_num) + '-(2*' + SD_cell + str(row_num) + ')'
        elif vendor_sum_i == 4: # -1SD
            formula = '=' + x_cell + str(row_num) + '-(1*' + SD_cell + str(row_num) + ')'
        elif vendor_sum_i == 5: # 1SD
            formula = '=' + x_cell + str(row_num) + '+(1*' + SD_cell + str(row_num) + ')'
        elif vendor_sum_i == 6: # 2SD
            formula = '=' + x_cell + str(row_num) + '+(2*' + SD_cell + str(row_num) + ')'
        elif vendor_sum_i == 7: # 3SD
            formula = '=' + x_cell + str(row_num) + '+(3*' + SD_cell + str(row_num) + ')'

        data_ws[cell] = formula
        data_ws[cell].border = thin_border
        vendor_sum_i = vendor_sum_i + 1

    row_num = row_num + 1
```
